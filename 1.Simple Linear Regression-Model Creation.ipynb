```python
#pandas is a library it contains functions and classes are in the database.all the function and classes are globalized in pandas.
import pandas as pd
```

```python
#In the dataset to call the function name and assign the parameter in dataset name.
dataset=pd.read_csv("Salary_Data.csv")
```

```python
dataset
```

```python
#In the dataset to seperate the column and it assign the variable name of input.
independent=dataset[["YearsExperience"]]
independent
```

```python
#In the dataset to seperate the column and it assign the variable name of output.
dependent=dataset[["Salary"]]
dependent
```

```python
#sklearn library is supports the Machine learning.
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(independent,dependent,test_size=0.30,random_state=0)#It
can load the function and then pass the parameter are assign in a variable name.
```

```python
#class name of linear regression are assign the variable name of regressor. 
from sklearn.linear_model import LinearRegression
regressor=LinearRegression()
regressor.fit(X_train,Y_train)#To load the fit method,to pass the parameter and it can be run the model to be created.
```

```python
weight=regressor.coef_#To find the weight(slope)of the created model.
weight
```

```python
bais=regressor.intercept_#To find the bais(intercept)of the created model.
bais
```

```python
y_pred=regressor.predict(X_test)#The model creation of the test data are assign in a variable.
```

```python
from sklearn.metrics import r2_score
r_score=r2_score(Y_test,y_pred)#To evaluate or check the test data are in a created model to pass the parameter is actual output and predicted output.
```

```python
r_score
```

```python
import pickle#Pickle library are used for save the created model.
filename="final_model_linaer.sav"#To assign file name in a variable.
```

```python
pickle.dump(regressor,open(filename,'wb'))# to save the created model should be write a new file.
```

```python
loaded_model=pickle.load(open("final_model_linaer.sav",'rb'))#Saved model to load the variable name and it assign the file name it can be read.
result=loaded_model.predict([[15]])#To check the input value.
```
